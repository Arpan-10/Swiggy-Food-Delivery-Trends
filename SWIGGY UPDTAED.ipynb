# Top 10 Cities Ordering Food from Swiggy

```python
top_10_cities = df['City'].value_counts().head(10).index
df_top_10_cities = df[df['City'].isin(top_10_cities)]
plt.figure(figsize=(10,6))
sns.countplot(x='City', data=df_top_10_cities, palette='viridis', order=top_10_cities)
plt.title('Top 10 Cities Ordering Food from Swiggy')
plt.xlabel('City')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Average Prices in Restaurants Across Cities

```python
city_avg_price = df.groupby('City')['Price'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=city_avg_price.values, y=city_avg_price.index, palette='viridis')
plt.title('Average Prices in Restaurants Across Cities')
plt.xlabel('Average Price')
plt.ylabel('City')
plt.show()
most_expensive_city = city_avg_price.idxmax()
print(f"The city with the most expensive food is: {most_expensive_city}")
```
